`Difference Between for and while`

| Feature      | for loop                            | while loop                       |
|--------------|-------------------------------------|----------------------------------|
| When used    | When we know how many times         | When we don’t know repetitions   |
| Syntax       | `for i in range():`                 | `while condition:`               |
| Example      | `for i in range(5)`                 | `while i < 5`                    |

`Set Operation`

| Operation             | Symbol / Method                    | Example                    | Result      |
|----------------------|------------------------------------|----------------------------|-------------|
| Union                | `|` or `.union()`                  | `{1,2,3} | {2,3,4}`        | `{1,2,3,4}` |

`Dictionary vs Set`

| Feature     | Dictionary (dict)            | Set (set)                               |
|------------|------------------------------|-----------------------------------------|
| Structure  | Key–Value pairs              | Unique values only                      |
| Ordered    |  Yes (since Python 3.7)     |  No                                    |
| Mutable    |  Yes                       |  Yes                                   |
| Duplicates |  Keys must be unique        |  No duplicates                         |
| Access     | By key                       | By membership (`in`)                    |
| Use Case   | Store related data           | Remove duplicates / perform set ops     |



## Error Handling

try:  
    # Code that might cause an error  
except:  
    # Code to run if error happens  


```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except:
    print("Something went wrong!")
```

```python
try:
    a = int(input("Enter number A: "))
    b = int(input("Enter number B: "))
    result = a / b
    print(result)

except ZeroDivisionError:
    print("You can't divide by zero!")

except ValueError:
    print("Please enter numbers only!")

except Exception as e:
    print("Unknown error:", e)
```

else

```python
try:
    number = int(input("Enter a number: "))
except ValueError:
    print("Invalid input!")
else:
    print("You entered:", number)
```

finally

```python
try:
    file = open("data.txt", "r")
    data = file.read()
    print(data)
except FileNotFoundError:
    print("File not found!")
finally:
    file.close()
    print("File closed.")
```

```python
age = int(input("Enter your age: "))
if age < 0:
    raise ValueError("Age cannot be negative!")
else:
    print("Your age is:", age)
```

```python
try:
    a = int(input("Enter number: "))
    try:
        print(10 / a)
    except ZeroDivisionError:
        print("Inner error: division by zero")
except ValueError:
    print("Outer error: invalid input")
```

| Exception         | Description                           |
|-------------------|---------------------------------------|
| ZeroDivisionError | Division by zero                      |
| ValueError        | Wrong data type (like `int("abc")`)   |
| FileNotFoundError | File doesn’t exist                    |
| TypeError         | Wrong type operation                  |
| IndexError        | Index out of range                    |
| KeyError          | Key not found in dictionary           |
| NameError         | Variable not defined                  |
| Exception         | Handles all types (general catch)     |
