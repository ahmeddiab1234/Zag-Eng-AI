##  1. Git Configuration

Used to set up your personal identity in Git.

```bash
git config --global user.name "ahmeddiab"      
git config --global user.email "shhth0034@gmail.com" 
git config user.name                            
git config user.email                          
git config --list                           
```


## 2. Git Basics
Essential commands to start a new Git project.
```bash

git init myproject              
cd myproject                  
git status                    
touch index.html           
git add index.html           
git commit -m "Create index.html"
git log                        
git add app.js index.html style.css
git commit -m "Add multiple files" 
```



##  3. Push & Pull
``` bash
Used to upload and download updates between local and remote repositories.

git push -u origin master    
git push                     
git pull origin master          
git fetch origin             
```


## 4. Undoing & Fixing Mistakes
Revert, reset, or correct commits and changes.

```bash
git reset --soft HEAD~1       
git reset --mixed HEAD~1    
git reset --hard HEAD~1    
git checkout <commit-hash>    
git checkout -- filename       
```



## 5. Branching & Merging
Work on multiple features simultaneously.

```bash
git branch           
git branch feature-branch    
git checkout feature-branch  
git checkout master    
git merge feature-branch      
git branch -d feature-branch    
git branch -D feature-branch  
```



## 6. Working with Remote Repositories
Link your local repo to GitHub and sync code.

```bash
git remote -v     
git remote add origin https://github.com/ahmeddiab1234/myproject.git
git clone https://github.com/ahmeddiab1234/myproject.git 
git remote remove origin      
```
